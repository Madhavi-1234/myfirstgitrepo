```python
import cv2
import matplotlib.pyplot as plt

img= cv2.imread("/content/WhatsApp Image 2025-09-08 at 10.15.52_67ef68a3.jpg")

img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

print(img_rgb)
print("shape of the array:", img_rgb.shape)

plt.imshow(img_rgb)
```
